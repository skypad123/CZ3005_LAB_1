```python
#print("All nodes : " + str(set(graph.keys())))
#print("--------------------------------")

#hospitals = hospitals
#nodesNotVisited = set(graph.keys()) - hospitals

#layer = 0
#result = { "0" : hospitals }
#temp = set()

#while nodesNotVisited != set():
#    layer += 1
#    print("layer: " + str(layer) + " to be Searched: " + str(result.get(str(layer-1))))
#    for layeritem in result.get(str(layer-1)):
#        
#        for x in graph.get(layeritem):
#            print("parent node: " + layeritem + " visited node: " + x)
#            if x in nodesNotVisited:
#                print("enter: " + x)
#                nodesNotVisited.remove(x) #add to visited
#                temp.update(x) #add to temp
#            print("nodes in nodes: " + str(nodesNotVisited))
#        result.update({str(layer): temp}) #add temp to result
#
#    print("current result: " + str(result))
#    print("--------------------------------")
#    temp = set()
    

def buildDistanceGraph(graph, nodes, layerLimit = -1,foundLimit = -1, findSpecific = {}) :
      
    nodesNotVisited = set(graph.keys()) - nodes 
    layer = 0
    foundCount = 0
    result = {"0": nodes}
    specificResult = {"0": nodes}

    while nodesNotVisited != set() and result.get(str(layer)) != set() and (layerLimit == -1 or layerLimit>layer):
        # moving the next layer
        layer += 1
        #setup set for new layer
        result.update({str(layer):set()})
        specificResult.update({str(layer):set()})
            
        # loop item in prev layer
        for layeritem in result.get(str(layer-1)):
            #loop to search new entries for current layer
            for x in graph.get(layeritem):
                ##result.update({str(layer):set()})

                #if x is not record before hand record that it been found and its distance
                if x in nodesNotVisited:
                    nodesNotVisited.remove(x)
                    result.get(str(layer)).update(x)

                    if findSpecific != {} and x in findSpecific:
                        specificResult.get(str(layer)).update(x)

                    if findSpecific == {} or x in findSpecific: 
                        foundCount+= 1

                    if foundLimit != -1 and foundCount >= foundLimit:
                        result.pop("0")
                        specificResult.pop("0")
                        return (specificResult if findSpecific != {} else result)
                    
    result.pop("0")
    specificResult.pop("0")
    return (specificResult if findSpecific != {} else result)

print()
nonHospitals = set(graph.keys()) - hospitals
print("All Non Hospital Nodes") 
print(nonHospitals)
print("Distance from Non Hospital Nodes to Closest Hospital") 
print(buildDistanceGraph(graph,hospitals))



nonHospitals = set(graph.keys()) - hospitals
print("All Non Hospital Nodes") 
print(nonHospitals)
for node in nonHospitals:
    print("Top "+ str(2) + " Hospital from  Node " + str(node))
    print(buildDistanceGraph(graph,{node},foundLimit=2,findSpecific=hospitals))

    
    
```
